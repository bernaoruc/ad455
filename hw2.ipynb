```r
##Make a select operation of columns, select()
## I will select the variables that I may use in the following steps
```

```r
infert %>% select(education,age, parity,stratum)
```

```r
##Make a filter operation of rows, filter()
##I filter women with 12+ years education and age between 25 to 30.
```

```r
infert %>%
    filter(education =="12+ yrs" & age %in% c(25:30))
```

```r
##Make a mutate operation to create a new column,  mutate()
## I will create a new column by using age variable, I will make 3 categories according to 1st and 3rd quartiles.
```

```r
summary(infert$age)
```

```r
infert_1 <- infert %>%
    mutate(age_grp = ifelse(age <= 28, "young", 
                    ifelse(age > 28 & age <= 35, "middle age", "older")))
head(infert_1)
```

```r
## Make a group_by and summarize operation , group_by() summarise()
## I want to see summarized table for average child owning for different age categories that I created.
```

```r
infert_1 %>%
    group_by(age_grp) %>%
    summarise(child = mean(parity))
```

```r
##Make a pivot_wider operation
## I created a summarized table of average, minimum, and maximum child for different age groups, and education level.
## then I will create a pivot table education/age group to see the avg child owning distribution. -> see infert_at_wide
```

```r
infert_2 <- infert_1 %>%
group_by(education, age_grp)%>%
select(education, age_grp, everything())%>%
summarise(avg_child= round(mean(parity),1),
          max_child=max(parity),
          min_child=min(parity))
infert_2
```

```r
infert_at_wide <- infert_2 %>% pivot_wider(id_cols = education, names_from = age_grp, values_from = avg_child)
infert_at_wide
```

```r
##Make a pivot_longer operation
### I make a longer pivot table of average child for education vs. age group
```

```r
infert_at_longer <- infert_at_wide %>% pivot_longer(cols = -"education",
                                                     names_to = "age_grp", 
                                                     values_to = "avg_child",
                                                     values_drop_na = T)
infert_at_longer
```

```r
##Make a left_join
##to do that I will create two sample data from my original dataset
## I joined the second data two first one on 3 columns = education, age_grp, and parity
```

```r
infert_s1 <- infert_1%>%
filter(education != "0-5yrs" & age_grp != "young")%>%
select(education, age_grp, parity )

infert_s2 <-  infert_1%>%
filter(education == "6-11yrs" & age_grp == "older")%>%
select(education, age_grp, spontaneous, parity)

infert_s1 %>% left_join(infert_s2, by = c("education", "age_grp", "parity"))
```
