```r
library(data.table)
library(datasets)
data()
```

```r
data(iris)
str(iris)
```

```r
iris <- as.data.table(iris)
str(iris)
```

```r
summary(iris)
```

```r
## i operation
##Tis filter retuns for lower values of the median of the 'Sepal.Length' and for higher values of the median of the 'Sepal.Width'
```

```r
iris[Sepal.Length < median(iris$Sepal.Length) &  Sepal.Width > median(iris$Sepal.Width)]
```

```r
##j operations
```

```r
iris[, .(Sepal.Length, Sepal.Width, Species,  Sep.Vol=(Sepal.Length*Sepal.Width))]
```

```r
## use ':=' notation
```

```r
iris[, Sep.Vol := Sepal.Length * Sepal.Width]
head(iris)
```

```r
## use 'by ' operation. I categorise by 'Species'
```

```r
iris[,.(sep.max_len = max(Sepal.Length, na.rm = T), sep.max_wid = max(Sepal.Width, na.rm = T) ), by = Species, ]
```

```r
filter.data <- iris[,.(sep.max_len = max(Sepal.Length, na.rm = T), sep.max_wid = max(Sepal.Width, na.rm = T) ), by = Species, ]
filter.data
```

```r
## use 'dcast()'
```

```r
species_volume <- dcast(filter.data, sep.max_wid ~ sep.max_len, value.var = "Species")
species_volume
```

```r
## use melt()
```

```r
melt(species_volume, id.vars = "sep.max_wid", variable.name = "sep.max_len", value.name = "Species")
```
